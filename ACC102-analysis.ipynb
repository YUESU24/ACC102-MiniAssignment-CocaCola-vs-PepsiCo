# Introduction

### Analytical Problem: A Comparative Analysis of Profitability Trends: Coca-Cola vs. PepsiCo
This project aims to compare the profitability of The Coca-Cola Company and PepsiCo over time. 
The main question is: Which company demonstrates stronger and more stable profitability?

### Target User
The intended users are investors and financial analysts who want to evaluate the financial performance of these two companies.

# Data Source

The dataset is obtained from WRDS (Wharton Research Data Services), which provides reliable financial data.
The dataset includes key financial variables such as:
- Net income (ni)
- Total assets (at)
- Shareholders’ equity (ceq)
- Sales (sale)
- Operating income (oiadp)

The data covers the period from 2015 to 2025 for Coca-Cola (KO) and PepsiCo (PEP).

The data were accessed and downloaded in April 2026.

# Data Processing

```python
import pandas as pd
```

```python
df = pd.read_csv("wrds_ko_pep_profitability.csv")
```

```python
df['datadate'] = pd.to_datetime(df['datadate'])
```

```python
df = df.sort_values(by=['tic', 'fyear'])
```

```python
df['ROA'] = df['ni'] / df['at']
df['ROE'] = df['ni'] / df['ceq']
df['Profit Margin'] = df['ni'] / df['sale']
df['Operating Margin'] = df['oiadp'] / df['sale']
```

```python
print(df[['tic', 'fyear', 'ROA', 'ROE', 'Profit Margin', 'Operating Margin']])
```

The dataset is cleaned and sorted by company and year. 
Key profitability ratios are calculated, including ROA, ROE, profit margin, and operating margin.

# Profitability Analysis & Visualization

```python
import matplotlib.pyplot as plt
```

```python
for company in df['tic'].unique():
    subset = df[df['tic'] == company]
    if company == 'KO':
        plt.plot(subset['fyear'], subset['ROA'], color='red', label='Coca-Cola')
    elif company == 'PEP':
        plt.plot(subset['fyear'], subset['ROA'], color='blue', label='PepsiCo')

plt.xlabel('Year')
plt.ylabel('ROA')
plt.title('ROA Comparison')
plt.legend()
plt.show()
```

### ROA Analysis

Coca-Cola shows a relatively stable ROA over most of the period, although it experiences a pronounced drop in 2017. After recovering from this dip, its ROA remains within a relatively narrow range, with only a modest decline around 2020. In contrast, PepsiCo’s ROA fluctuates more noticeably, particularly due to a sharp peak in 2018, after which it returns to a lower and more stable level.

This suggests that Coca-Cola uses its assets more consistently to generate profit, while PepsiCo’s asset efficiency is more variable over time.

```python
for company in df['tic'].unique():
    subset = df[df['tic'] == company]
    if company == 'KO':
        plt.plot(subset['fyear'], subset['ROE'], color='red', label='Coca-Cola')
    elif company == 'PEP':
        plt.plot(subset['fyear'], subset['ROE'], color='blue', label='PepsiCo')

plt.xlabel('Year')
plt.ylabel('ROE')
plt.title('ROE Comparison')
plt.legend()
plt.show()
```

### ROE Analysis

PepsiCo maintains a higher ROE than Coca-Cola for most of the observed period, although Coca-Cola slightly exceeds PepsiCo in 2025. PepsiCo’s ROE also shows greater volatility, with a particularly sharp spike in 2018. In contrast, Coca-Cola’s ROE remains within a more stable range throughout the period.

These results indicate that PepsiCo often delivers higher returns to shareholders, but with greater variability, whereas Coca-Cola provides more stable returns over time.

```python
for company in df['tic'].unique():
    subset = df[df['tic'] == company]
    if company == 'KO':
        plt.plot(subset['fyear'], subset['Profit Margin'], color='red', label='Coca-Cola')
    elif company == 'PEP':
        plt.plot(subset['fyear'], subset['Profit Margin'], color='blue', label='PepsiCo')

plt.xlabel('Year')
plt.ylabel('Profit Margin')
plt.title('Profit Margin Comparison')
plt.legend()
plt.show()
```

### Profit Margin Analysis

Coca-Cola maintains a higher profit margin than PepsiCo for most of the observed period, except for 2017 when its margin temporarily falls below that of PepsiCo. After this sharp decline, Coca-Cola’s profit margin recovers quickly in 2018 and remains well above PepsiCo in subsequent years. PepsiCo’s profit margin, by contrast, remains lower and fluctuates within a relatively narrow range after 2019.

This suggests that Coca-Cola generally has a stronger ability to convert revenue into net income, while PepsiCo shows more limited improvement in overall profitability.

```python
for company in df['tic'].unique():
    subset = df[df['tic'] == company]
    if company == 'KO':
        plt.plot(subset['fyear'], subset['Operating Margin'], color='red', label='Coca-Cola')
    elif company == 'PEP':
        plt.plot(subset['fyear'], subset['Operating Margin'], color='blue', label='PepsiCo')

plt.xlabel('Year')
plt.ylabel('Operating Margin')
plt.title('Operating Margin Comparison')
plt.legend()
plt.show()
```

### Operating Margin Analysis

Coca-Cola consistently reports a higher operating margin than PepsiCo across the entire period. Its operating margin rises noticeably from 2015 to 2018 and then remains at a relatively high level with moderate fluctuations. PepsiCo’s operating margin shows a mild increase in the early years, followed by a decline from 2018 to 2022 and a modest recovery thereafter.

These results indicate that Coca-Cola maintains stronger operating profitability, while PepsiCo experiences more variation in its core operating performance.

## Overall Comparison

Overall, Coca-Cola performs better than PepsiCo in ROA, profit margin, and operating margin, while PepsiCo records a higher but more volatile ROE for most years. Across the four indicators, Coca-Cola demonstrates more consistent profitability, whereas PepsiCo shows stronger returns in certain periods but with greater fluctuations.
